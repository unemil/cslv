```python
import cv2
from PIL import Image
```

```python
img = cv2.imread('tests/captcha.png')

Image.fromarray(img)
```

```python
img = cv2.resize(img, (img.shape[1] * 2, img.shape[0] * 2), interpolation=cv2.INTER_LINEAR)

cv2.imwrite('tests/captcha_resize.png', img)
Image.fromarray(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imwrite('tests/captcha_grayscale.png', img)
Image.fromarray(img)
```

```python
img = cv2.bilateralFilter(img, 5, 15, 15)

cv2.imwrite('tests/captcha_filter.png', img)
Image.fromarray(img)
```

```python
(T, img) = cv2.threshold(img, 200, 255, (cv2.THRESH_BINARY))

cv2.imwrite('tests/captcha_threshold.png', img)
Image.fromarray(img)
```
