```python
import cv2
from PIL import Image
```

```python
img = cv2.imread('tests/captcha.png')

Image.fromarray(img)
```

```python
img = cv2.resize(img, (img.shape[1] * 2, img.shape[0] * 2), interpolation=cv2.INTER_LINEAR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.bilateralFilter(img, 5, 15, 15)

Image.fromarray(img)
```

```python
(T, img) = cv2.threshold(img, 210, 255, (cv2.THRESH_BINARY))

Image.fromarray(img)
```
